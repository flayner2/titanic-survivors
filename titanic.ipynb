## Predictions

```python
from sklearn import linear_model
```

### Creating dummies for the categorical data

```python
features = ['Pclass', 'Sex', 'Embarked']
cat_features = features[1:]

X_train, X_test = training_data[features], test_data[features]

for feature in cat_features:
    temp_train = pd.get_dummies(training_data[feature])
    temp_test = pd.get_dummies(test_data[feature])
    X_train, X_test = X_train.join(temp_train), X_test.join(temp_test)

X_train, X_test = X_train.drop(cat_features, axis=1), X_test.drop(cat_features, axis=1)
```

```python
y_train = training_data['Survived']
```

```python
X_train
```

```python
y_train
```

### Implement the model

```python
model = linear_model.LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

### Predict the values for the test set

```python
X_test
```

```python
y_test = model.predict(X_test)
```

```python
y_test
```

```python
test_data['Survived'] = y_test
test_data
```

### Plotting the predictions
