## Exploratory analysis

```python
p = sns.countplot(data=training_data, x='Survived', hue='Pclass', palette='Set2')
plt.title('Plot of Survived by Class', weight='bold')
plt.legend(bbox_to_anchor=(1.02, 0.5), loc=2, borderaxespad=0.)
plt.ylabel('Count')
```

```python
p = sns.countplot(data=training_data, x='Pclass', hue='Survived', palette='Set2')
plt.title('Plot of Class by Survived', weight='bold')
plt.legend(bbox_to_anchor=(1.02, 0.5), loc=2, borderaxespad=0.)
plt.ylabel('Count')
```

```python
p = sns.countplot(data=training_data, x='Survived', hue='Sex', palette='Set2')
plt.title('Plot of Survived by Sex', weight='bold')
plt.legend(bbox_to_anchor=(1.02, 0.5), loc=2, borderaxespad=0.)
plt.ylabel('Count')
```

```python
p = sns.countplot(data=training_data, x='Sex', hue='Survived', palette='Set2')
plt.title('Plot of Sex by Survived', weight='bold')
plt.legend(bbox_to_anchor=(1.02, 0.5), loc=2, borderaxespad=0.)
plt.ylabel('Count')
```

```python
p = sns.catplot(data=training_data, x='Pclass', col='Survived', hue='Sex', kind='count', palette='Set2')
plt.suptitle('Count of passengers that died or survived based on class and sex', weight='bold')
plt.subplots_adjust(top=0.85)
p.set_axis_labels('Passenger class', 'Count')
```
